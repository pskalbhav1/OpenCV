```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img=cv.imread('images.jpg')
mask=np.zeros(img.shape[:2],np.uint8)
bgdModel=np.zeros((1,65),np.float64)
fgdModel=np.zeros((1,65),np.float64)
rect=(80,20,200,180)
rect1=(80,50,130,180)
rect2=(140,20,200,180)
cv.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
mask2=np.where((mask==2) | (mask==0),0,1).astype('uint8')
img=img*mask2[:,:,np.newaxis]
cv.grabCut(img,mask,rect1,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
mask3=np.where((mask==2) | (mask==0),0,1).astype('uint8')
img1=img*mask3[:,:,np.newaxis]
cv.grabCut(img,mask,rect2,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
mask4=np.where((mask==2) | (mask==0),0,1).astype('uint8')
img2=img*mask4[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()
plt.imshow(img1)
plt.colorbar()
plt.show()
plt.imshow(img2)
plt.colorbar()
plt.show()
```
